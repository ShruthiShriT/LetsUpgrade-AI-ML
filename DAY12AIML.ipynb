```python
#manwhitneytest
data1=data.loc[data['Attrition']=='Yes']

data1.to_csv("Attrition_yes")
data2=data.loc[data['Attrition']=='No']
data2.to_csv("Attrition_no")
data2
```

```python
from scipy.stats import mannwhitneyu
```

```python
stats,p=mannwhitneyu(data1.Education,data1.Education)
print(stats,p)
```

```python
stats,p=mannwhitneyu(data1.Age,data1.Age)
print(stats,p)
```

```python
stats,p=mannwhitneyu(data1.DistanceFromHome,data1.DistanceFromHome)
print(stats,p)
```

```python
stats,p=mannwhitneyu(data1.NumCompaniesWorked,data1.NumCompaniesWorked)
print(stats,p)
```

```python
stats,p=mannwhitneyu(data1.PercentSalaryHike,data1.PercentSalaryHike)
print(stats,p)
```

```python
stats,p=mannwhitneyu(data1.StockOptionLevel,data1.StockOptionLevel)
print(stats,p)
```

```python
stats,p=mannwhitneyu(data1.TotalWorkingYears,data1.TotalWorkingYears)
print(stats,p)
```

```python
stats,p=mannwhitneyu(data1.TrainingTimesLastYear,data1.TrainingTimesLastYear)
print(stats,p)
```

```python

stats,p=mannwhitneyu(data1.YearsAtCompany,data1.YearsAtCompany)
print(stats,p)
```

null hypothesis:there is no siginificant difference between Attrition no and Attrition yes set of Age,DistanceFromHome,Education,NumCompaniesWorked,PercentSalaryHike,StockOptionLevel,TotalWorkingYears,TrainingTimesLastYear,YearsAtCompany
alternate hypothesis:there is siginificant difference between Attrition no and Attrition yes set of Age,DistanceFromHome,Education,NumCompaniesWorked,PercentSalaryHike,StockOptionLevel,TotalWorkingYears,TrainingTimesLastYear,YearsAtCompany
conclusion:
    p>0.05 we can accept the null hypothesis
