```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Bank_Personal_Loan_Modelling.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['ZIP Code','ID'], axis=1)
```

```python
y=df.PersonalLoan
```

```python
x=df[["Age","Experience","Income","Family","CCAvg","Education","Mortgage","Securities Account","CD Account","Online","CreditCard"]]
```

```python
import statsmodels.api as sm
```

```python
x1=sm.add_constant(x)
```

```python
log=sm.Logit(y,x1)
```

```python
result=log.fit()
```

```python
result.summary()
```

# the p values should be less than 0.05 so that we can say they are significantly  important in the above indepedent variables 

Age,Experience,Mortgage >0.05 so they are not needed as that are not significantly important then we can have values as 

Income,Family,Education,Online,CreditCard,CCAvg,Securities Account <0.05 they more significantly important to determine the result of 

the personal loan and we can use the coefficeints to form the equation and get the p value to determine the personal loan

```python
df=pd.read_csv("general_data.csv")
```

```python
df.head()
```

```python
from sklearn import preprocessing
```

```python
new_NumCompaniesWorked=np.where(df["NumCompaniesWorked"].isnull(),2,df["NumCompaniesWorked"])
df["NumCompaniesWorked"]=new_NumCompaniesWorked
new_TotalWorkingYears=np.where(df["TotalWorkingYears"].isnull(),11.0,df["TotalWorkingYears"])
df["TotalWorkingYears"]=new_TotalWorkingYears
```

```python
label_encoder=preprocessing.LabelEncoder()
df["BusinessTravel"]=label_encoder.fit_transform(df["BusinessTravel"])
df["Attrition"]=label_encoder.fit_transform(df["Attrition"])
df["Department"]=label_encoder.fit_transform(df["Department"])
df["EducationField"]=label_encoder.fit_transform(df["EducationField"])
df["Gender"]=label_encoder.fit_transform(df["Gender"])
df["JobRole"]=label_encoder.fit_transform(df["JobRole"])
df["MaritalStatus"]=label_encoder.fit_transform(df["MaritalStatus"])
```

```python
df.head(2)
```

```python
df.drop(['BusinessTravel','EmployeeCount','EmployeeID','Over18','StandardHours'], axis=1)
```

```python
y=df.Attrition
```

```python
x=df[["Age","Department","DistanceFromHome","Education","EducationField","Gender","NumCompaniesWorked","PercentSalaryHike","StockOptionLevel","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany","YearsSinceLastPromotion","YearsWithCurrManager"]]
```

```python
x1=sm.add_constant(x)
x1.isnull().sum()
```

```python
df.isnull().sum()
```

```python
log=sm.Logit(y,x1)
```

```python
# df.replace([np.inf, -np.inf], np.nan)
```

```python
result=log.fit()
```

```python
result.summary()
```
